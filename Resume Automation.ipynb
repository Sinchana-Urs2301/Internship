```python
import pandas as pd
import re
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("C:/Users/sinch/Downloads/DS_Projects/job_descriptions.csv" )
df
```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()   # lowercase
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # remove symbols
    
    words = text.split()
    words = [w for w in words if w not in stop_words]
    
    return " ".join(words)
print(stop_words)
```

```python
data['clean_resume'] = data['resume_text'].apply(clean_text)

data[['resume_text', 'clean_resume']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(data['clean_resume'])
X
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(X)
X
```
