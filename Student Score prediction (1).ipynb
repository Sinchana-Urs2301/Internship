# Model performance

```python
X = data['hours_studied'].values
Y = data['final_score'].values

mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)

numer = 0
denom = 0

for i in range(n):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2

m = numer/denom
c = mean_y -(m * mean_x)

print(f"m:{m}")
print(f"c:{c}")
```

```python
ss_tot = 0
ss_res = 0

for i in range(n):
    y_pred = (m * X[i]) + c
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2

r2 = 1 - (ss_res/ss_tot)
print(f"r2 = {r2}")
```

### R2 Indicates ~45% of varience explained

```python
y_true = np.array([Y1])
y_pred = np.array([y_pred])

rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
print(rmse)
```
