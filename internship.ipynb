# Showing Hours of studies and Final score through scatter plot

```python
plt.scatter(data['hours_studied'],data['final_score'])
plt.show()
```

```python
X = data['hours_studied'].values
Y = data['final_score'].values
```

# Calculating for linear regression and executive through graph

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)

numer = 0
denom = 0

for i in range(n):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2

m = numer/denom
c = mean_y -(m * mean_x)

print(f"m:{m}")
print(f"c:{c}")
```

```python
max_x = np.max(X)+1
min_x = np.min(X)-1

x = np.linspace(min_x , max_x)
print(x)
print(Y)
y = (m * x) + c

plt.plot(x,y,color='darkblue' , label='Linear Regression' , linestyle='--')
plt.scatter(X , Y , color='red' , label='Linear Regression' , linestyle='--')
plt.title('Student Performance Prediction')
plt.xlabel('Hour Studied')
plt.ylabel('Exam Score')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

# Calculating R2

```python
ss_tot = 0
ss_res = 0

for i in range(n):
    y_pred = (m * X[i]) + c
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2

r2 = 1 - (ss_res/ss_tot)
print(f"r2 = {r2}")
```
